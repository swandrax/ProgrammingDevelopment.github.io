```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
x1 = np.linspace(0, 10, 100)

# create a plot figure
fig = plt.figure()
plt.plot(x1, np.tan(x1), '-')
plt.plot(x1, np.cos(x1), '--');
```

Visualisasi Data

---
Menggunakan visualisasi data menggunakan Pyplot sangatlah mudah,Nilai sumbu x berkisar antara 0-3 dan sumbu y dari 1-4. Jika kita memberikan sebuah daftar atau array pada perintah plot(), matplotlib mengasumsikan bahwa daftar atau array tersebut adalah sebuah urutan nilai y, dan secara otomatis menghasilkan nilai x. Karena index python dimulai dengan 0, vektor x default memiliki panjang yang sama dengan y tetapi dimulai dengan 0. Oleh karena itu, data x adalah [0,1,2,3] dan data y adalah [1,2,3,4].









```python
plt.plot([1, 2, 3, 4])
plt.ylabel('Angka Rata - Rata')
plt.show()
```


Plot() => adalah perintah serbaguna
---

Perintah Serbaguna yang bertujuan untuk menerima beberapa argumen. contohnya untuk memplotkan x terhadap y, kita dapat mengeluarkan perintah berikut:

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

Interface menggunakan State-Machine
---
Pyplot menyediakan interface Library Pustaka berorientasi yang mendasari secara implisit dan otomatis membuat gambar untuk mencapai plot yang membuat otomatis membuat sumbu dan garis dan mencapai plot yang di inginkan

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='Penjualan')
plt.plot(x, x**2, label='pembelian')
plt.plot(x, x**3, label='investasi toko')

plt.xlabel('Label x')
plt.ylabel('label y')

plt.title("grafik sederhana market-bola")

plt.legend()

plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from scipy.stats import ttest_ind

# Generate a synthetic dataset
X, y = make_blobs(n_samples=300, centers=3, random_state=42)

# Data Preparation
data = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
data['Target'] = y

# Data Visualization
plt.scatter(data['Feature1'], data['Feature2'], c=data['Target'], cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Synthetic Dataset with 3 Clusters')
plt.show()

# Machine Learning Model
X_train, X_test, y_train, y_test = train_test_split(data[['Feature1', 'Feature2']], data['Target'], test_size=0.2, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f"Accuracy of Random Forest Classifier: {accuracy}")

# Statistical Analysis
group1_data = data[data['Target'] == 0]['Feature1']
group2_data = data[data['Target'] == 1]['Feature1']

result = ttest_ind(group1_data, group2_data)
print(f"T-statistic: {result.statistic}, p-value: {result.pvalue}")

# Clustering
kmeans = KMeans(n_clusters=3)
data['Cluster'] = kmeans.fit_predict(data[['Feature1', 'Feature2']])

# Visualization of Clusters
plt.scatter(data['Feature1'], data['Feature2'], c=data['Cluster'], cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Cluster Visualization with K-Means')
plt.show()
```
